# **Introdução**

O desemprego é um fenômeno social e econômico que afeta sociedades em todo o mundo, manifestando-se como a ausência de trabalho formal ou informal para indivíduos aptos e dispostos a trabalhar. Ele é um dos principais indicadores da saúde econômica de um país e reflete a realidade de milhões de pessoas, influenciando diretamente o bem-estar coletivo, a renda, o consumo, a arrecadação e o desenvolvimento humano. Do ponto de vista social, o desemprego pode gerar exclusão, desigualdade e aumento da violência. Economicamente, compromete a capacidade de crescimento, reduzindo a produção, a atração de investimentos e a arrecadação tributária. Estudos sobre desemprego e criminalidade indicam que a ausência de oportunidades formais pode impulsionar práticas ilícitas, agravando a vulnerabilidade social. Compreender o desemprego em suas múltiplas dimensões é fundamental para avaliar a saúde econômica e orientar políticas públicas capazes de mitigar impactos negativos, promovendo desenvolvimento sustentável e inclusão social.

# **Referencial Teórico**

Principais Conceitos

O desemprego é o conceito central desta análise. Segundo o Instituto Brasileiro de Geografia e Estatística (IBGE, 2024), o desemprego é definido como a situação de indivíduos em idade ativa que, apesar de estarem disponíveis e aptos para o trabalho, não encontram ocupação no mercado formal ou informal. Esse indicador é amplamente utilizado para avaliar a saúde da economia de um país.

Trabalhos Correlacionados

O estudo do desemprego tem sido objeto central da economia do trabalho e das ciências sociais, dada sua relação direta com bem-estar social, crescimento econômico e segurança pública.

Do ponto de vista social, a literatura aponta que a ausência de emprego contribui para a exclusão social, a degradação do capital humano e o aumento da criminalidade. Kahn (2020), em estudo para o Fórum Brasileiro de Segurança Pública, destaca que as taxas de violência estão fortemente correlacionadas com o desemprego, especialmente em contextos de vulnerabilidade social. De maneira semelhante, Santiago e Guimarães (2017) analisaram séries temporais e identificaram padrões consistentes de aumento da criminalidade em períodos de maior desocupação.

Na dimensão econômica, Lopes, Cavalcante e Oliveira (2020) reforçam que o desemprego provoca redução do consumo, queda da arrecadação tributária e comprometimento da capacidade de investimento público e privado. Esses efeitos, quando persistentes, ampliam desigualdades e travam o desenvolvimento sustentável.

A interseção entre desigualdade racial, gênero e escolaridade também é amplamente documentada. Pesquisas do IBGE mostram que mulheres e pessoas negras enfrentam taxas de desemprego superiores à média nacional, além de apresentarem menores rendimentos mesmo quando inseridos no mercado de trabalho (IBGE, 2024).

Em síntese, o referencial teórico demonstra que o desemprego não é apenas um indicador econômico, mas um fenômeno multidimensional, cujos impactos sociais e econômicos se reforçam mutuamente. Esse entendimento fundamenta a presente pesquisa, que busca, por meio da análise dos dados da PNAD Contínua, oferecer subsídios técnicos e sociais para a formulação de políticas públicas mais eficazes.
