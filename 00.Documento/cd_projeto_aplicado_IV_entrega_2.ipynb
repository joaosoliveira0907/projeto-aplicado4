# **Introdução**

O desemprego é um fenômeno social e econômico que afeta sociedades em todo o
mundo, manifestando-se como a ausência de trabalho formal ou informal para
indivíduos aptos e dispostos a trabalhar. Ele é um dos principais indicadores da
saúde econômica de um país e reflete a realidade de milhões de pessoas,
influenciando diretamente o bem-estar coletivo, a renda, o consumo, a arrecadação
e o desenvolvimento humano.
Do ponto de vista social, o desemprego pode gerar exclusão, desigualdade e
aumento da violência. Economicamente, compromete a capacidade de crescimento,
reduzindo a produção, a atração de investimentos e a arrecadação tributária.
Estudos sobre desemprego e criminalidade indicam que a ausência de
oportunidades formais pode impulsionar práticas ilícitas, agravando a vulnerabilidade
social.
Compreender o desemprego em suas múltiplas dimensões é fundamental para
avaliar a saúde econômica e orientar políticas públicas capazes de mitigar impactos
negativos, promovendo desenvolvimento sustentável e inclusão social.

# **Referencial Teórico**

O estudo do desemprego tem sido objeto central da economia do trabalho e das
ciências sociais, dada sua relação direta com bem-estar social, crescimento
econômico e segurança pública.
Segundo o IBGE (2024), o desemprego é definido como a situação de indivíduos em
idade ativa que, apesar de estarem disponíveis e aptos para o trabalho, não
encontram ocupação no mercado formal ou informal. Esse indicador é amplamente
utilizado para avaliar a saúde da economia de um país.
Do ponto de vista social, a literatura aponta que a ausência de emprego contribui
para a exclusão social, a degradação do capital humano e o aumento da
criminalidade. Kahn (2020), em estudo para o Fórum Brasileiro de Segurança
Pública, destaca que as taxas de violência estão fortemente correlacionadas com o
desemprego, especialmente em contextos de vulnerabilidade social. De maneira
semelhante, Santiago e Guimarães (2017) analisaram séries temporais e
identificaram padrões consistentes de aumento da criminalidade em períodos de
maior desocupação.
Na dimensão econômica, Lopes, Cavalcante e Oliveira (2020) reforçam que o
desemprego provoca redução do consumo, queda da arrecadação tributária e
comprometimento da capacidade de investimento público e privado. Esses efeitos,
quando persistentes, ampliam desigualdades e travam o desenvolvimento
sustentável.
A interseção entre desigualdade racial, gênero e escolaridade também é
amplamente documentada. Pesquisas do IBGE mostram que mulheres e pessoas
negras enfrentam taxas de desemprego superiores à média nacional, além de

apresentarem menores rendimentos mesmo quando inseridos no mercado de
trabalho. Essa dimensão interseccional reforça a necessidade de análises que vão
além dos números agregados, evidenciando vulnerabilidades específicas que
precisam ser enfrentadas por políticas públicas focalizadas.
Em síntese, o referencial teórico demonstra que o desemprego não é apenas um
indicador econômico, mas um fenômeno multidimensional, cujos impactos sociais e
econômicos se reforçam mutuamente. Esse entendimento fundamenta a presente
pesquisa, que busca, por meio da análise dos dados da PNAD Contínua, oferecer
subsídios técnicos e sociais para a formulação de políticas públicas mais eficazes.

# **Resultados e Discussão**

Os resultados reforçam que o desemprego no Brasil não é homogêneo, mas varia
significativamente entre grupos sociais. Constata-se que mulheres, pessoas negras
e indivíduos com baixa escolaridade são desproporcionalmente afetados, tanto em
termos de taxas de desemprego quanto em rendimento médio.

crise (como 2015–2016 e a pandemia de 2020), seguido de recuperações parciais.

não são suficientes para reduzir disparidades específicas.
Essa constatação sugere a necessidade de políticas públicas focalizadas, com
atenção especial a grupos historicamente vulneráveis.
